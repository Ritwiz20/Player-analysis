```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("Virat_Kohli.csv")
```

```python
df.head()
```

```python
# no of matches he played at different positions

positions = df["Pos"].unique()
# positions

# formatting position columns
df["Pos"] = df["Pos"].map({
     1.0 : "Batting at 1",
     2.0 : "Batting at 2",
     3.0 : "Batting at 3",
     4.0 : "Batting at 4",
     5.0 : "Batting at 5",
     6.0 : "Batting at 6",
     7.0 : "Batting at 7"
    })

pos_counts = df["Pos"].value_counts() # frequency of occurence of a single element in a col
# print(pos_counts)

pos_counts_values = pos_counts.values
pos_counts_labels = pos_counts.index
# print(pos_count_values)

fig = plt.figure(figsize = [5,5])
plt.pie(pos_counts_values, labels = pos_counts_labels)
plt.show()
```

```python

def show_pie_plot(df, key):
    counts = df[key].value_counts()
    count_values = counts.values
    count_label = counts.index

    fig = plt.figure(figsize=[10,7])
    plt.pie(count_values, labels=count_label)
    plt.show()
```

```python
""" No of matches at different positions """

show_pie_plot(df,"Pos")
```

```python
"""" No of matches in different grounds """

show_pie_plot(df,"Ground")
```

```python
"""" No of matches against different opposition """

show_pie_plot(df, "Opposition")
```

```python
""" Total 6s at diff positions """

sixes_at_pos = df.groupby("Pos")["6s"].sum()
print(sixes_at_pos)

sixes_at_pos_value = sixes_at_pos.values
sixes_at_pos_labels = sixes_at_pos.index

fig = plt.figure(figsize=[10,7])
plt.bar(
    sixes_at_pos_labels,
    sixes_at_pos_value,
    color="green",
    width=0.3,
    
)
plt.xlabel("Playing at position ")
plt.ylabel("Number of 6s")
plt.show()
```

```python
""" Total runs scored against diff countries """

runs = df.groupby("Opposition")["Runs"].sum()

my_value = runs.values
my_label = runs.index

fig = plt.figure(figsize=[17,10])
plt.bar(
    my_label,
    my_value,
    color="#ff99cc",
    width=0.3,
    
)
plt.xlabel("Opposition")
plt.ylabel("Runs")
plt.show()
```

```python
""" No of centuries in 1st and 2nd innings """

centuries = df.query("Runs >= 100")
centuries

innings = centuries["Inns"].value_counts()
innings

my_value = innings.values
my_label = innings.index

fig = plt.figure(figsize=[10,7])
plt.bar(
    my_label,
    my_value,
    color="#99ccff",
    width=0.3,
    
)
plt.xlabel("Innings")
plt.ylabel("Century Count")
plt.show()
```
